```python
import pathlib

folder_path = pathlib.Path(r"C:\Users\User\Documents\שנה ג\data_science\data-science\govData")
# print files mame in folder
files = folder_path.glob("*.csv")
for file in files:  
    print(file.name)
```

```python
files_to_rename = [
    "ride_data.csv",
    "ride_data2.csv",
    "ride_data3.csv",
    "ride_data4.csv",
    "ride_data_merged.csv"
]

rename_columns = {
    "_id": "record_id",
    "month": "month",
    "route_id": "route_id",
    "DayOfWeek": "day_of_week",
    "HourSourceTime": "scheduled_departure_time",
    "StopSequence_Rishui": "stop_sequence",
    "StopCode": "stop_code",
    "count_common": "n_observations",
    "timeCumSum_mean": "mean_cumulative_travel_time_min",
    "timeCumSum_std": "std_cumulative_travel_time_min",
    "distCumSum_mean": "mean_cumulative_distance_m",
    "distCumSum_std": "std_cumulative_distance_m"
}
chosen_route_ids_by_line = {
    "37936": 15,
    "5499": 22,
    "10802": 19,
    "10398": 17
}
```

```python
for file_name in files_to_rename:
    file_path = folder_path / file_name
    df = pd.read_csv(file_path)
    df.rename(columns=rename_columns, inplace=True)
    df["route_name"] = df["route_id"].astype(str).map(chosen_route_ids_by_line) # add route_name column based on route_id


    output_file = folder_path / f"renamed_{file_name}"
    df.to_csv(output_file, index=False)
    print(f"Renamed columns in {file_name} and saved as {output_file.name}")
```
